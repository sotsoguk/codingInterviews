# 2. Backtracking

## 2.1 N-Queens

```python
def printBoard(placement,n):
    for i in range(0,n):
        for j in range(0,n):
            if placement[i] == j:
                print('X',end='')
            else:
                print('.',end='')
        print('')
```

```python
a=[3,6,2,7,1,4,0,5]
```

```python
printBoard(a,8)
```

```python
boardSize = 8
board = [0] * boardSize
```

```python
def placeQueens(board,r)  :
    n = len(board)
    if r>=n:
        printBoard(board,n)
        print("")
    else:
        for j in range(0,n):
            legal = True
            for i in range(0,r):
                if board[i] == j or board[i] == j+r-i or board[i] == j-r+i:
                    legal = False
            if legal == True:
                board[r] = j
                placeQueens(board,r+1)
```

```python
placeQueens(board,0)
```
