## Unnecessary Work

Print the sum of all positive integers a,b,c,d <1000, so that a^3+b^3=c^3+d^3

### Naive Approach O(n^4)

```python
n = 1001

for a in range(1,n):
    for b in range(1,n):
        for c in range(1,n):
            for d in range(1,n):
                #if pow(a,3)+pow(b,3) == pow(c,3) + pow(d,3):
                if a**3 + b**3 == c**3 + d**3:
                    print(a,b,c,d)
                    # if d is found, no other value can be computed
                    break
```

For any a,b,c there is only one d=(a^3+b^3-c^3)^(1/3)

```python
n = 1001

for a in range(1,n):
    for b in range(1,n):
        for c in range(1,n):
            d = int(pow(abs(a**3+b**3-c**3),1/3))
            if a**3 + b**3 == c**3 + d**3:
                print(a,b,c,d)
```

### A smarter solution

Store the sum of a^3 + b^3 in a dictionary.

```python
from collections import defaultdict
n = 1001
pairs = defaultdict(list)
for a in range(1,n):
    for b in range(a,n):
        #print(a,b)
        pairs[a**3+b**3].append([a,b])
```

```python
elems = 0
for k,v in pairs.items():
    
    if len(v)>=2:
        print(k,v)
    
print(elems)
```
